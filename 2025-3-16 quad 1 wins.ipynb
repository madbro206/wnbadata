```python
import pandas as pd
import matplotlib.pyplot as plt
from io import StringIO
from matplotlib.ticker import MaxNLocator
```

```python
csv_data = """school,quad_1_wins
South Carolina,16
Texas,14
UCLA,13
USC,13
Duke,8
Maryland,8
North Carolina State,8
TCU,8
North Carolina,7
Oklahoma,7
Iowa,6
LSU,6
Notre Dame,6
Ole Miss,6
Tennessee,6
Alabama,5
Connecticut,5
Kentucky,5
Michigan,5
Ohio State,5
Oklahoma State,5
Illinois,4
Indiana,4
Michigan State,4
Nebraska,4
Oregon,4
Baylor,3
California,3
Florida State,3
Mississippi State,3
Utah,3
Vanderbilt,3
Washington,3
Colorado,2
Harvard,2
Kansas State,2
Louisville,2
Saint Joseph's,2
Texas A&M,2
West Virginia,2
Arizona,1
Columbia,1
Florida,1
Georgetown,1
Georgia,1
Georgia Tech,1
Iowa State,1
Kansas,1
Penn State,1
Princeton,1
Richmond,1
South Florida,1
Stanford,1
Texas Tech,1
Virginia,1
Virginia Tech,1
Wisconsin,1"""

# Use StringIO to simulate reading from a file-like object.
csv_file = StringIO(csv_data)

# Read the CSV into a Pandas DataFrame.
df = pd.read_csv(csv_file)
```

```python
df_sorted = df.sort_values(by='quad_1_wins', ascending=False)

plt.figure(figsize=(7, 11))
bars = plt.barh(df_sorted['school'], df_sorted['quad_1_wins'], color='skyblue')

# Ensure x-axis shows integers only.
plt.gca().xaxis.set_major_locator(MaxNLocator(integer=True))

# Add labels to the end of each bar.
for bar in bars:
    plt.text(bar.get_width() + 0.5, bar.get_y() + bar.get_height() / 2,
             f'{bar.get_width():.0f}', va='center', ha='left')

# Customize the chart.
plt.xlabel('Quad 1 Wins')
plt.ylabel('School')
plt.title('Quad 1 Wins by School, Teams with >0 only')
plt.gca().invert_yaxis()  # Invert y-axis to have the highest value at the top.
plt.tight_layout()
plt.show()
```
